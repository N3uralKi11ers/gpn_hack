Туториал как использовать функции из tools

```python
import pandas as pd
```

```python
df = pd.read_parquet("../data/end_df.parquet")
df.head()
```

```python
def compare_prices_1_2(
    df: pd.DataFrame, 
    product_title: str, 
    place_title: str, 
    competitor: str,
):
    df_filtered = df.copy()
    df_filtered = df_filtered.loc[
        (df_filtered['product'] == product_title)
        & (df_filtered['place'] == place_title)
    ]
    
    our_prices = df_filtered['Our_company']
    competitor_prices = df_filtered[competitor]
    result = (our_prices < competitor_prices * 1.2).all()
    return result
```

Валидация на $price_{our} < price_{competitor} * 1.2$ для всех продуктов, мест и конкурентов

```python


results = []
for product in df['product'].unique():
  for place in df['place'].unique():
    for competitor in ['Арториас&Co', 'Длань господня', 'ЛилIT', 'Светлые Души']:
      res = compare_prices_1_2(df, product, place, competitor)
      results.append([res, product, place, competitor])
      
results
```

```python
def compare_prices_1(
    df: pd.DataFrame, 
    product_title: str, 
    place_title: str
):
    df_filtered = df.loc[
        (df['product'] == product_title)
        & (df['place'] == place_title)
    ].copy()
    df_filtered['date'] = pd.to_datetime(df_filtered[['year', 'month', 'day']])
    df_filtered.sort_values('date')
    
    _prices = df_filtered['Our_company']
    prev = None
    flag = True
    
    for _price in _prices:
        if prev is not None and _price - prev > 1:
            flag = False
                
        prev = _price
            
    return flag
```

```python
compare_prices_1(df, 'Целебные травы', 'Врата Балдура')
```

```python
results = []
for product in df['product'].unique():
  for place in df['place'].unique():
    res = compare_prices_1(df, product, place)
    results.append([res, product, place])

results
```
